```python
from nltk import bigrams

nltk.download('punkt')

# Load your dataset (e.g., a CSV file)
df = data

all_in = []

# Let's use 'description' column to generate bigrams
for index, row in df.iterrows():
    headline_text = row['headline_text']

    # Tokenize the description
    text_tokens = nltk.word_tokenize(headline_text)

    # Generate bigrams
    text_bigrams = list(bigrams(text_tokens))

    print(f"Headline Text: {headline_text}\nBigrams: {text_bigrams}\n")
    all_in.append(text_bigrams)
    #flatten(all_in)
```

```python
flattened_bigrams = [bigram for sublist in all_in for bigram in sublist]

print(flattened_bigrams)
```

```python
from collections import Counter
# Get frequency of bigrams
bigram_freqs = Counter(flattened_bigrams)

# Print bigrams, sorted by frequency
for bigram, freq in bigram_freqs.most_common():
  if freq > 10:
    print(f"Bigram: {bigram}, Frequency: {freq}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Download stopwords and WordNet corpus
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('punkt')
# nltk.download('omw-1.4')

# Load stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Initialize WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to filter the text
def filter_text(text):
    if isinstance(text, float):
        return ""
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stopwords and punctuation
    filtered_tokens = [token for token in tokens if token not in stop_words and token not in punctuation]

    # Lemmatize the tokens
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    filtered_text = ' '.join(lemmatized_tokens)
    
    return filtered_text
```

```python
# Apply the filter_text function to the "headline_text" column
data['filtered_text'] = data['headline_text'].apply(filter_text)
```

```python
import spacy

# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')

# Function to perform NER
def perform_ner(text):
    if pd.isnull(text):
        return []  # Return an empty list if the value is missing
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities
```

```python
data.loc[:, 'Entities'] = data['headline_text'].apply(perform_ner)
```

```python
# Count the entities by category
category_counts = defaultdict(int)

for entities in data['Entities']:
    for entity in entities:
        category = entity[1]
        category_counts[category] += 1

# Sort the categories and counts in ascending order
sorted_categories = sorted(category_counts.keys(), key=lambda x: category_counts[x], reverse=True)
sorted_counts = [category_counts[category] for category in sorted_categories]


# Plot the distribution
plt.bar(sorted_categories, sorted_counts)
plt.xlabel('Entity Category')
plt.ylabel('Count')
plt.title('Distribution of Entities by Category (Ascending Order)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame based on the desired year range
filtered_df = data[(data['year'] >= 2015) & (data['year'] <= 2020)]

# Flatten the list of entities within the filtered DataFrame
all_entities = [entity for entities in filtered_df['Entities'] for entity in entities]

# Count the occurrences of each entity by category
entity_counts = Counter(entity[1] for entity in all_entities)

# Get unique entity categories
categories = list(set(entity[1] for entity in all_entities))

# Initialize a dictionary to store the category counts over time
category_counts_by_year = {category: [] for category in categories}

# Iterate over the filtered DataFrame to count entities by category per year
for year in range(2015, 2021):
    year_df = filtered_df[filtered_df['year'] == year]
    year_entities = [entity[1] for entities in year_df['Entities'] for entity in entities]
    year_entity_counts = Counter(year_entities)
    
    for category in categories:
        category_counts_by_year[category].append(year_entity_counts[category])

# Plot the trend of entity frequencies by category over time
years = range(2015, 2021)
plt.figure(figsize=(10, 6))

for category in categories:
    plt.plot(years, category_counts_by_year[category], label=category)

plt.xlabel('Year')
plt.ylabel('Entity Count')
plt.title('Trend of Entity Frequencies by Category (2015-2020)')
plt.legend()
plt.show()
```

```python
df_copy = data.copy()

# Split the tuple in 'Entities' column and extract the first word
df_copy['Entities'] = df_copy['Entities'].apply(lambda x: [entity[0] for entity in x])

# Group by 'headline_category' and 'Entities', and count the occurrences
grouped_entities = df_copy.explode('Entities').groupby(['headline_category', 'Entities']).size().reset_index(name='Count')

# Sort the entities within each category by count in descending order
sorted_entities = grouped_entities.groupby('headline_category').apply(lambda x: x.nlargest(5, 'Count')).reset_index(drop=True)

# Generate the plot
plt.figure(figsize=(12, 8))
for category, df_copy in sorted_entities.groupby('headline_category'):
    entities = df_copy['Entities']
    counts = df_copy['Count']
    plt.bar(entities, counts, label=category)

plt.xlabel('Entities')
plt.ylabel('Count')
plt.title('Top 5 Entities in Each Category')
plt.legend()
plt.tight_layout()
plt.show()
```
